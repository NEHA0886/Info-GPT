```python
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
model = T5ForConditionalGeneration.from_pretrained("/content/drive/MyDrive/flan-t5-base-prompt-response", device_map="auto")
```

```python
questions = ["What were the key areas of emphasis for Tata Steel's profitability?",
 ' How did Tata Steel manage its financial capital and debt?',
 ' How did Tata Steel manage cash flows during the pandemic?',
 ' How did Tata Steel reduce its debt levels over time?',
 ' What is the core focus of innovation at Tata Steel?',
 'What are the strategic linkages that Tata Steel aims to establish?',
 ' What is the purpose of the Graphene business at Tata Steel?',
 'What is the Shikhar25 program at Tata Steel?',
 ' How does Tata Steel invest in augmenting its capacities?',
 " What is Tata Steel's approach to steel manufacturing?",
 " What are the key minerals that Tata Steel's mines produce?",
 " What is the capacity of Tata Steel's India crude steel production?",
 " What is the significance of Tata Steel's integrated operations?",
 ' How does Tata Steel ensure cost-competitiveness in global markets?',
 ' How does Tata Steel monetize in-house developed Intellectual Property?',
 " How did Tata Steel's Graphene business perform in terms of revenue?",
 ' What were the consolidated financial performance highlights of Tata Steel?']
```

```python
answers = []
from tqdm import tqdm

for q in tqdm(questions):
  input_text = f"Answer the following Question:\n{q.strip()}\nAnswer:\n"
  input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")
  outputs = model.generate(input_ids, max_new_tokens=500, do_sample=True,
            top_k=50, top_p=0.95, temperature=0.3, repetition_penalty=1.2)
  answers.append(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
